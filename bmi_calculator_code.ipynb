```python
import cv2
from matplotlib import pyplot as plt
import def_Information

class BMI:
    
    #เมนูเริ่มต้น
    def Start():
        print('-'*117)    
        print('')
        get_ipython().run_cell_magic('HTML', '', '<center>\n<span style="color: black; font-family: supermarket; font-size: large;">\n<center>\nโปรแกรมคำนวณค่า BMI\n<br/>\nของ\n<br/>\n</center>\n</span>\n<MARQUEE behavior=alternate direction=left scrollAmount=3 width="10%">\n<font face=Webdings>\n4\n</font>\n</MARQUEE>\n<MARQUEE scrollAmount=1 direction=left width="2%">\n...\n</MARQUEE>\n<span style="color: black; font-family: supermarket; font-size: large;">\nนักเรียนโรงเรียนวิทยาศาสตร์จุฬาภรณราชวิทยาลัย ปทุมธานี\n</span>\n<MARQUEE scrollAmount=1 direction=right width="2%">\n...\n</MARQUEE>\n<MARQUEE behavior=alternate direction=right scrollAmount=3 width="10%">\n<font face=Webdings>\n3\n</font>\n</MARQUEE>\n<hr>\n<img src="https://upload.wikimedia.org/wikipedia/th/thumb/f/f0/Pccnewlogo.png/200px-Pccnewlogo.png" width="100px" height="100px">\n<hr>\n</center>')   
        print('-'*117)

    #เปิดดูข้อมูลเกี่ยวกับ BMI
    def ImportInformation():
        def_Information.Information() 

    #วิธีการใช้งานโปรแกรม
    def Instruction():
        print('')    
        print('\tวิธีการใช้งานโปรแกรม')
        print('')
        print('กรุณาป้อนข้อมูลน้ำหนักในหน่วย กิโลกรัม')
        print('กรุณาป้อนข้อมูลส่วนสูงในหน่วย เซนติเมตร')
        print('')
        print('กรอกข้อมูลต่อไป เพื่อคำนวณคนต่อไป')
        print('กรอก 0 ในช่่องน้ำหนักและส่วนสูง เมื่อต้องการจะหยุด')
        print('')

    #การใส่รูปภาพเกณฑ์ BMI        
    def Picture():
        x = cv2.imread('bmipic.jpg')
        plt.imshow(x[10:1500,5:1500,::-1])

    #การนำข้อมูลเข้า + คำนวณ + แสดงผลลัพธ์    
    def CalBMI():
        data = []
        mea  = []

        while True:
            weight = float(input('ป้อนน้ำหนัก: '))
            height = float(input('ป้อนส่วนสูง: '))
            print('')
            print('คนต่อไป: ')

            if weight == 0 and height == 0:
                print('')
                print('กรุณารอสักครู่...')
                print('')
                print('-'*117)
                print('')
                break        
            else:
                BMI = float('%.2f'%(weight/((height/100)**2)))
                data.append(BMI)
                try:
                    if BMI <= 18.5:
                        mea.append('น้อยกว่าเกณฑ์')
                    elif 22.9 >= BMI >= 18.5:
                        mea.append('สมส่วน')
                    elif 24.9 >= BMI >= 23.0:
                        mea.append('น้ำหนักเกิน')
                    elif 29.9 >= BMI >= 25.0:
                        mea.append('โรคอ้วน')
                    elif 24.9 >= BMI >= 23.0:
                        mea.append('โรคอ้วนอันตราย')
                except IndexError :
                    print('ข้อมูลผิดพลาด')

        print('\t\tรายงานผล')     
        print('')

        for i in range (len(data)):
            print('BMI ของคนที่ {} {} อยู่ในเกณฑ์: {}'.format(i+1, data[i], mea[i]))
            print('')

    #รายชื่อผู้จัดทำ
    def Member(): 
        x = open('รายชื่อผู้จัดทำ.txt','w')
        x.write('จัดทำโดย นักเรียนชั้น ม.4/3\n\nนายธัญชนก วัตถาภรณ์ศิริ (6)\nนายวรันธร จันทร์สว่าง (9)\nนางสาวปาณิศา จิรถาวรกุล (20)')
        x.close()

    #พิมพ์ข้อเสนอแนะ
    def Note():
        y = open('ข้อเสนอแนะ.txt','w')
        z = input('ข้อเสนอแนะ: ')
        y.write('ข้อเสนอแนะ: '+str(z))
        y.close()

    Start()       
    ImportInformation()
    Instruction()
    Picture()
    CalBMI()  
    Member()
    Note()
```
